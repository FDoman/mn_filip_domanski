Używając do tego funkcji: [ylim](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.ylim.html), [xlim](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xlim.html), [axis](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.axis.html)

```python
# inicjalizacja wektora z zad 1
x = np.linspace(-10, 10, 1000)
y = x**3 - 3 * x

#1
plt.plot(x, y)
# ograniczenie argumentów do pożądanego przedziału
plt.xlim(-1, 1)
# zmiana przedziału wartości w celu pokazania całego wykresu
plt.ylim(-2, 2)

plt.title("f(x) = x^3 - 3x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()

#2
plt.plot(x, y)
plt.xlim(-10, -1)
plt.ylim(-1000, 100)

plt.title("f(x) = x^3 - 3x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()

#3
plt.plot(x,y)
# zmiana przedziałów argumentów oraz wartości za pomocą jednej funkcji
plt.axis([1, 10, -100, 1000])

plt.title("f(x) = x^3 - 3x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

```python
# zamiana zadanych parametrów na podstawowe jednostki
m = 2500 * 0.001
v = 60 / 3.6

# obliczenie ciepła w dżulach
Q = (m*v**2)/2
print('Wartość wydzielanego ciepła dla masy równej 2.5 kg oraz prędkości równej 60 km/h:', Q, 'J')

# obliczenie ciepła w kilokaloriach
Q1 = Q/(4.1855*1000)
print('Ta sama wartość w kilokaloriach:', Q1, 'kcal')

# inicjalizacja nowych parametrów oraz wektora ciepła
m1 = 3
v1 = np.arange(0, 200.1, 0.1)
q = (m1*v1**2)/2

# wykres w skali liniowej
plt.plot(v1, q)
plt.xlim(200, 0)
plt.title('Zmiana ilości wydzielanego ciepła w procesie hamowania')
plt.xlabel('prędkość v [m/s]')
plt.ylabel('ciepło Q [J]')
plt.show()

# wykres w skali półlogarytmicznej
plt.semilogy(v1, q)
plt.xlim(200, 0)
plt.title('Zmiana ilości wydzielanego ciepła w procesie hamowania\n oś OY w skali logarytmicznej')
plt.xlabel('prędkość v [m/s]')
plt.ylabel('ciepło Q [J]')
plt.show()
```

**Wniosek** - w pierwszej części zadania policzone zostało wydzielone ciepło w procesie hamowania z wcześniejszą konwersją wartości na podstawowe jednostki układu SI. Wynik podany został najpierw w dżulach, a następnie w kilokaloriach.
W drugiej części wyrysowano dwa wykresy obrazujące zmianę ciepła dla ciała o zadanej masie i przedziale prędkości: w skali liniowej i półlogarytmicznej za pomocą funkcji $semilogy$. Aby wyświetlić wykresy z malejącymi argumentami v: <200, 0>, użyto funkcji $xlim$.
Wykresy poprawnie wyświetlają ilość wydzielanego ciepła, obrazując jego spadek wraz ze zmniejszaniem się prędkości.

Częstą praktyką szczególnie w dziedzinie analizy danych, statystyce, uczeniu maszynowym, itp. jest tworzenie na podstawie biblioteki [Matplotlib](https://matplotlib.org/) własnych szablonów wykresów stworzonych na podstawie funkcji. Najprostszy przykład to przeładowanie funkcji plot tak by wykres miał czerwony kolor:

```python
def my_plot(x,y,xlabel,ylabel,title,label):
    if x.shape != y.shape or  min(x.shape)==0:
        return None
    fig, ax = plt.subplots()
    ax.plot(x, y, 'r', label=label)
    ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    ax.legend()
    return fig

x = np.linspace(-10,10)
# inicjalizacja wektora f(x) = sin(x)
y = np.sin(x)
my_plot(x,y,'x','y','Funkcja sin', 'sin(x)')
```

***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
# inicjalizacja wektora kąta oraz parametrów
fi = np.linspace(0, 20, 1000)
a1 = 100
a2 = 0.01
b1 = 0.4
b2 = 0.1

# inicjalizacja wektorów w sposób parametryczny
r1 = a1 * np.exp(b1 * fi)
x1 = r1 * np.cos(fi)
y1 = r1 * np.sin(fi)

r2 = a2 * np.exp(b2 * fi)
x2 = r2 * np.cos(fi)
y2 = r2 * np.sin(fi)

# wykreślenie wykresów w dwóch kolumnach
main.parallel_plot(x1, y1, x2, y2, 'x', 'y', 'x', 'y', 'Spirala logarytmiczna', '|')
plt.show()
```

**Wniosek** - funkcja $parallel\_plot$, zaimplementowana w poprzednim zadaniu, została użyta do wykreślenia spiral logarytmicznych na dwóch sąsiadujących wykresach. W tym celu zdefiniowano wektor opisujący kąt oraz parametry a i b, gdzie a jest zmieniane i odpowiada za wielkość krzywej, natomiast b jest stałe i odpowiada za "szybkość" zakręcania spirali. Następnie zapisano wektory x i y, za pomocą równań parametrycznych w celu wykreślenia funkcji w układzie współrzędnych kartezjańskich.

Pierwsza spirala określona jest na szerokim przedziale (duża wartość parametru a), ale wolno 'zakręca' dążąc do półprostej (większy współczynnik b), natomiast druga wykreślona jest w okolicy zera (mała wartość parametru a), ale szybko 'zakręca' dążąc do okręgu (mniejszy współczynnik b).
